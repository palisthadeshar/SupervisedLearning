```python
#here x represents age of car and y represents its speed
import matplotlib.pyplot as plt
```

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
#importing scipy and drawing the line of regression
from scipy import stats
```

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
#method that returns some important key value of linear regression
slope, intercept, r, p, std_err = stats.linregress(x,y)
```

```python
#create a new function that uses slope and intercept to return new value
def myfunc(x):
    return slope * x + intercept
```

```python
#run each values of the x array through the function. This will result in a new array with values for y-axis
mymodel = list(map(myfunc,x))
```

```python
plt.scatter(x,y) #draw the original scatter plot
plt.plot(x,mymodel) #draws the line of linear regression
plt.show()
```

```python
#r = coefficient of correlation
#its value ranges from -1 to 1 where 0 means no relationship, 1and -1 means 100% realted. This value will be calculated by python and scipy
#all we have to do si feed it with x and y values.
```
