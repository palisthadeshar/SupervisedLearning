Data cleaning and feature engineering

```python
df_alexa
```

```python
df_alexa = df_alexa.drop(['date','rating'], axis=1)
```

```python
df_alexa
```

```python
#translate variations into numbers using get_dummies(encoding in a way)
variation_dummies = pd.get_dummies(df_alexa['variation'], drop_first =True)
```

```python
variation_dummies
```

```python
df_alexa.drop('variation', axis=1, inplace=True)
```

```python
#add the encoded column of variations
df_alexa = pd.concat([df_alexa, variation_dummies], axis=1)
```

```python
df_alexa
```

```python
#change every single reviews into number
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
alexa_countvectorizer = vectorizer.fit_transform(df_alexa['verified_reviews'])#counts how manyn words are there in the reviews
```

```python
alexa_countvectorizer.shape 
```

```python
print(vectorizer.get_feature_names_out())
```

```python
encoded_reviews = pd.DataFrame(alexa_countvectorizer.toarray())
```

```python
df_alexa.drop('verified_reviews', axis=1,inplace=True)
```

```python
df_alexa = pd.concat([df_alexa, encoded_reviews], axis = 1)
```

```python
df_alexa
```

```python
X = df_alexa.drop(['feedback'], axis = 1)
```

```python
X.shape
```

```python
y = df_alexa['feedback']
```

```python
y.shape
```

Model training

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
randomforest_classifier = RandomForestClassifier(n_estimators = 100, criterion = 'entropy')
randomforest_classifier.fit(X_train, y_train)
```

Evaluate the model

```python
y_predict_train = randomforest_classifier.predict(X_train)
cm = confusion_matrix(y_train, y_predict_train)
```

```python
y_train
```

```python
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_train, y_predict_train))
```

```python
y_predict_test = randomforest_classifier.predict(X_test)
cm = confusion_matrix(y_test, y_predict_test)
```

```python
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict_test))
```
