#model training

```python
X = cancer_df.drop(['target'], axis = 1)
```

```python
X
```

```python
y = cancer_df['target']
```

```python
y
```

Divide data into training and test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 5)
```

```python
X_train.shape
```

```python
y_train.shape
```

Training the data

```python
#importing class
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

```python
svc_model = SVC() #instantiating the object out of the class
```

```python
svc_model.fit(X_train, y_train) #trained the model
```

Evaluating the model

```python
y_predict = svc_model.predict(X_test)
```

```python
y_predict
```

```python
cm  = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot =True)
plt.show()
```

```python
print(classification_report(y_test, y_predict))
```

noramalizing the model - improving the model through linearlization

```python
min_train = X_train.min()
```

```python
min_train
```

```python
range_train = (X_train - min_train).max()
```

```python
range_train
```

```python
X_train_scaled = (X_train - min_train)/range_train #scaling data(0 to 1)
```

```python
X_train_scaled
```

```python
sns.scatterplot(x = X_train['mean area'], y = X_train['mean smoothness'], hue = y_train)
plt.show()
```

```python
sns.scatterplot(x = X_train_scaled['mean area'], y = X_train_scaled['mean smoothness'], hue = y_train)
plt.show()
```

```python
#for testing data set
min_test = X_test.min()
```

```python
range_test = (X_test - min_test).max()
```

```python
range_test
```

```python
X_test_scaled = (X_test - min_test)/range_test #scaling data(0 to 1)
```

```python
X_test_scaled
```

```python
sns.scatterplot(x = X_test_scaled['mean area'], y = X_test_scaled['mean smoothness'], hue = y_test)
plt.show()
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train_scaled, y_train)
```

```python
y_predict = svc_model.predict(X_test_scaled)
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot = True, fmt = "d")
plt.show()
```

```python
print(classification_report(y_test, y_predict))
```

Parameter optimization-improving the model

```python
param_grid = {'C':[0.1, 1, 10, 100], 'gamma':[1,0.1,0.01,0.001], 'kernel':['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 4)
```

```python
grid.fit(X_train_scaled, y_train)
```

```python
grid.best_params_
```

```python
grid_predict = grid.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test, grid_predict)
```

```python
sns.heatmap(cm, annot =True)
plt.show()
```

```python
print(classification_report(y_test, y_predict))
```
